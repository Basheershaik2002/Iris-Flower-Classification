```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Separate the dataset by species
setosa = iris[iris['species'] == 'setosa']
versicolor = iris[iris['species'] == 'versicolor']
virginica = iris[iris['species'] == 'virginica']

# Plot distributions for each species separately with distplot
sns.distplot(setosa['sepal_length'], bins=15, kde=True, color='blue', label='Setosa')
sns.distplot(versicolor['sepal_length'], bins=15, kde=True, color='green', label='Versicolor')
sns.distplot(virginica['sepal_length'], bins=15, kde=True, color='red', label='Virginica')

plt.title('Distribution of Sepal Length by Species')
plt.legend()

plt.show()
```
